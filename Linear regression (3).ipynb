#### Fit a Linear regression model

```python
# Build regression model

import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
```

```python
model1.summary()
```

```python
#Plot the scatter plot and overlay the fitted straight line using matplotlip
x = data["daily"].values
y = data["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s = 30)
b0 = 13.84
b1 = 1.33
#predicted response vector
y_hat = b0 + b1*x
# plotting the regression line
plt.plot(x, y_hat, color = "g")
# putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Plot the  linear regression line using seaborn regplot() method
sns.regplot(x="daily", y="sunday", data=data)
plt.xlim([0,1250])
plt.show()
```

#### Predict for new data points(test data)

```python
#Predict sunday sales for 200 and 300 and 1500 daily circulation
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
# Predict on all given training data

pred = model1.predict(data["daily"])
pred
```

```python
# Add predicted values as a column in data1
data["Y_hat"] = pred
data
```

```python
# Compute the error values (residuals) and add as another column
data["residuals"]=data["sunday"]-data["Y_hat"]
data
```

```python
# Compute Mean Squared error for the model

mse = np.mean((data["daily"]-data["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
# Compute Mean Absolute Error (MAE)

mae = np.mean(np.abs(data["daily"]-data["Y_hat"]))
mae
```

### Checking the model rediuals scatter plot(for homoscedasticity) and normality
- check the scatter plot for residuals Vs Y_hat
- Check the Q-Q plot residuals distribution
- Check the kde hist plot residuals

```python
# Plot the residuals versus y_hat ( to check wheather residuals are independent of each other)
plt.scatter(data["Y_hat"],data["residuals"])
```

#### Observations
- There appears to be no trend and the residuals are randomly placed around the zero error line
- Hence the assumption of homoscedasticty (constant variance in residuals) is satisfied


```python
#Plot the Q-Q plot (to check the normality of residuals)
import statsmodels.api as sm
sm.qqplot(data["residuals"], line='45', fit=True)
plt.show()
```

```python
# Plot the kde distribution for residuals
sns.histplot(data["residuals"], kde =True)
```

#### Observations
- The data points are seen to closely follow the reference line of normality
- Hence the residuals are approximately normally distributed as also can be seen from the kde distribution
- The assumptioms of simple linear regressionn are verified and found to be valid in this case.
